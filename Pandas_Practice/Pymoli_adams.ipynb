```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Total players summary table
total_players = purchase_data.SN.nunique()
playercount_table = pd.DataFrame()
playercount_table["Total Players"] = [total_players]

playercount_table
```

```python
#Purchasing Analysis
#Total number items purchased
item_total = purchase_data["Item ID"].nunique()
#Average price
avg_price = purchase_data["Price"].mean()
#Total purchases
purchase_total = len(purchase_data)
#Total revenue
rev_total = purchase_data["Price"].sum()
#Summary Table
purchasing_table = pd.DataFrame()
purchasing_table["Number of Unique Items"] = [item_total]
purchasing_table["Average Price"] = [avg_price]  
purchasing_table["Number of Purchases"] = [purchase_total]
purchasing_table["Total Revenue"] = [rev_total]
purchasing_table["Average Price"] = purchasing_table["Average Price"].map("${0:,.2f}".format)
purchasing_table["Total Revenue"] = purchasing_table["Total Revenue"].map("${0:,.2f}".format)
purchasing_table
```

```python
#Gender Demographics
demographics = purchase_data.loc[:, ["SN", "Gender", "Age", "Price"]]
demographics = demographics.drop_duplicates()
count_gender = demographics["Gender"].value_counts()
count_gender
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed 
perc_count = (count_gender/total_players)
perc_count
#Summary Chart
gender_table = pd.DataFrame()
gender_table["Gender Count"] = count_gender
gender_table["Percentage of Gender"] = perc_count
gender_table["Percentage of Gender"] = gender_table["Percentage of Gender"].map("{:,.2%}".format)
gender_table
```

```python
#Purchasing Analysis

#Purchase Count by gender
purchase_countgen = purchase_data.groupby("Gender").count()["Price"]
#Average Purchase Price by gender
price_gender = purchase_data.groupby("Gender").mean()["Price"]
#Total Purchase Value  by gender
total_purchasegen = purchase_data.groupby("Gender").sum()["Price"]
#Average Purchase Total per Person by Gender
avg_gender = total_purchasegen / count_gender

purchasing_table = pd.DataFrame()
purchasing_table["Purchase Count"] = purchase_countgen
purchasing_table["Avg Purchase Price"] = price_gender
purchasing_table["Total Purchase Value"] = total_purchasegen
purchasing_table["Avg Purchase Total"] = avg_gender
purchasing_table["Avg Purchase Price"] = purchasing_table["Avg Purchase Price"].map("${0:,.2f}".format)
purchasing_table["Total Purchase Value"] = purchasing_table["Total Purchase Value"].map("${0:,.2f}".format)
purchasing_table["Avg Purchase Total"] = purchasing_table["Avg Purchase Total"].map("${0:,.2f}".format)
purchasing_table
```

```python
#Age Demographics The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
age_bin = [0, 9.999, 14.999, 19.999, 24.999, 29.999, 34.999, 39.999, 99999]
labels = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40 plus"]
demographics["Age Ranges"] = pd.cut(demographics["Age"], age_bin, labels=labels)
demographics
#Purchase Count & Percentage of purchase by age
count_age = demographics["Age Ranges"].value_counts()
perc_age = (count_age/ total_players)

#Summary Table
age_demographics = pd.DataFrame()
age_demographics["Count of Players"] = count_age
age_demographics["Percentage of Players"] = perc_age

#Percentage formatting provided by program tutor- reformatted bottom to top of code
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:,.2%}".format)
age_demographics = age_demographics.sort_index()
age_demographics
```

```python
#Determine Average Purchase Price, Total Purchase Value, Average Purchase Total per Person by Age Group
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bin, labels=labels)
age_app = demographics.groupby("Age Ranges").mean()["Price"]
age_tpv = demographics.groupby("Age Ranges").sum()["Price"]
age_apt = age_tpv / count_age

#Summary Table
age_table = pd.DataFrame()
age_table["Avg Purchase Price"] = age_app
age_table["Total Purchase Value"] = age_tpv
age_table["Avg Purchase Total"] = age_apt
age_table["Total Purchase Value"] = age_table["Total Purchase Value"].map("${0:,.2f}".format)
age_table["Avg Purchase Price"] = age_table["Avg Purchase Price"].map("${0:,.2f}".format)
age_table["Avg Purchase Total"] = age_table["Avg Purchase Total"].map("${0:,.2f}".format)
age_table
```

```python
#Top Spenders- Identify the the top 5 spenders in the game by total purchase value, 
#Table: SN, Purchase Count, Average Purchase Price, Total Purchase Value
sn_tpv = purchase_data.groupby(["SN"]).sum()["Price"]
sn_pc = purchase_data.groupby(["SN"]).count()["Price"]
sn_app = purchase_data.groupby(["SN"]).mean()["Price"]

sn_data = pd.DataFrame()
sn_data["Total Purchase Value"] = sn_tpv
sn_data["Purchase Count"] = sn_pc
sn_data["Average Purchase Price"] = sn_app
sn_data["Total Purchase Value"] = sn_data["Total Purchase Value"].map("${0:,.2f}".format)
sn_data["Average Purchase Price"] = sn_data["Average Purchase Price"].map("${0:,.2f}".format)
sn_sort= sn_data.sort_values("Total Purchase Value", ascending=False)
sn_sort.head(5)
```

```python
#Most Popular Items- Identify the 5 most popular items by purchase count
#Table: Item ID, Item Name, Purchase Count, Item Price, Total Purchase Value
items = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]] 
item_count = items.groupby(["Item ID", "Item Name"]).count()["Price"]
item_tpv = items.groupby(["Item ID", "Item Name"]).sum()["Price"]
item_tp = items.groupby(["Item ID", "Item Name"]).mean()["Price"]

item_data = pd.DataFrame()
item_data["Total Purchase Count"] = item_count
item_data["Item Price"] = item_tp
item_data["Total Purchase Value"] = item_tpv
#formatting from https://www.thiscodeworks.com/ & program tutor 
#Went from bottom of code to top reformatting
item_data["Item Price"] = item_data["Item Price"].map("${0:,.2f}".format)
item_data["Total Purchase Value"] = item_data["Total Purchase Value"].map("${0:,.2f}".format)

item_data = item_data.sort_values("Total Purchase Count", ascending=False)
item_data.head(5)
```
